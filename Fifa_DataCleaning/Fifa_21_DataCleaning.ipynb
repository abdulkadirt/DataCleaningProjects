```python
for index , row in df.iterrows() :
    if 'On Loan' in row["Contract"] or 'Free' in row["Contract"]:
        print(row["Contract"])
```

```python
def extract_contract_info(contract):
    if contract == "Free" or "On Loan" in contract :
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else :
        start_date , end_date = contract.split(" ~ ")
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
        
    return start_date , end_date , contract_length

# Contract sütununa uygulanınca 3 yeni kolon oluşturur.

new_cols = ["Contract_Start","Contract_End","Contract_Length"]
new_df = df["Contract"].apply(lambda x : pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    df.insert(loc=df.columns.get_loc("Contract") + 1 + i , column=new_cols[i] , value=new_df[i])
```

yukarıda oluşturduğum yeni kolonları hazırdaki df e ekledim. özellikle Contract kolonunun yanına eklemek istediğim için bi tık karışık olmuş olaiblir.

```python
# Kontratları kategorilere göre ayırabiliriz.

def categorize_contract_status(contract):
    if contract == "Free" :
        return "Free"
    elif "On Loan" in contract:
        return 'On Loan'
    else :
        return "Contract"
    
# kontrat kategorisi sütununu ekleme
df.insert(loc = df.columns.get_loc("Contract_Length") + 1 ,
          column= "Contrat_Status" ,
          value=df["Contract"].apply(categorize_contract_status))
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else: 
        feet , inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
    
# height sğtununa bu fonksiyonu uygulamak gerekiyor.
df["Height"] = df["Height"].apply(convert_height)
```

```python
print(df["Height"].unique()) # hepsi int artık.
print(df["Height"].dtypes)
```

```python
df = df.rename(columns={"Height" : "Height(cm)"}) # tamamen cm cinsinden değerler içeriyor artık.
```

```python
def convert_weight(weight) :
    if "kg" in weight:
        return int(weight.strip("kg"))
    else :
        pounds = int(weight.strip("lbs"))
        return round(pounds / 2.205)
    
    
# kolonu fonksiyonu uyguluyoruz :)
df["Weight"] = df["Weight"].apply(convert_weight)
```

```python
df["Weight"].unique() # tip int oldu ve her şey uygun gzüküyor artık.
```

```python
df = df.rename(columns={"Weight" : "Weight(kg)"}) 
# tüm ağırlıklar kg olarak belirtildiği için kolon ismini ona göre güncelliyorum.
```
