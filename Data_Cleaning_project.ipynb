```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Cricket_test_data.csv",sep = ";")
df.head(10)
```

Verinin içerdiği sütunlar kriket oyuncularının performans istatistiklerine dair verileri temsil ediyor. İşte sütunların anlamları:

1. **Span**: Oyuncunun kariyerindeki aktif yıllar (örneğin 1928-1948).
2. **Mat**: Oynanan toplam maç sayısı (Matches).
3. **Inns**: Oyuncunun yer aldığı vuruş (innings) sayısı.
4. **NO**: Not Out, yani oyuncunun maç boyunca elenmeden kalma sayısı.
5. **Runs**: Toplam koşu (run) sayısı.
6. **HS**: En yüksek skor (High Score). "*" işareti, elenmeden kalındığını gösterir.
7. **Ave**: Ortalama koşu sayısı (Average), toplam koşuların oynanan vuruşlara bölünmesiyle elde edilir.
8. **BF**: Ball Faced, yani oyuncunun karşı karşıya kaldığı toplam top sayısı.
9. **SR**: Vuruş oranı (Strike Rate), her 100 topta yapılan koşu sayısını ifade eder.
10. **100**: Yüz sayı ve üzeri koşu yapılan maç sayısı (Centuries).
11. **50**: Elli sayı ve üzeri koşu yapılan maç sayısı (Half-Centuries).
12. **0**: Sıfırda elenilen vuruş (innings) sayısı (Ducks).
13. **4s**: Vuruşla alınan dört koşu (boundary) sayısı.
14. **6s**: Vuruşla alınan altı koşu (six) sayısı.
15. **Column15**: Görünüşe göre bir ek sütun olabilir; içeriği tam belli değil, ama muhtemelen bir hata ya da açıklanmamış bir veri.

###### rename multiple columns in a list.

Sütunların isimlerini değiştirerek başalıyorum. Kriketle alakalı her şeyi bilmediğim için başta chatgpt ye sormuştum ama her seferinde dönüp not okumak istemiyorum. Ayrıca fark edilebileceği üzere sütun isimlerinde kaydırma yapılmış. Kendim düzenlerken yapmışım onu düzeltmek gerekiyor.

###### Çok önemli not : NaN değerler float olarak kabul ediliyor. Ondan dolayı int'e çevirirken eğer bu satırda NaN değer varsa sorun yaşarsın.

```python
df.isna().any().sum()
```

```python
df['4s'] = df['4s'].astype(int)
df['6s'] = df['6s'].astype(int) # bu da sayısal olsa daha iyi olur gibi.
df['Balls_Faced'] = df['Balls_Faced'].astype(int)
df['Hihgest_Inns_Score'] = df['Hihgest_Inns_Score'].astype(int)
```

```python
df.dtypes
```

sonradan eklediğim tüm sütunları str metotlarıyla oluşturduğum için object oalrak gözüküyorlar ancak o kısım da düzeltilmeli.

```python
df = df.astype({"Start_year" : int , "Final_year" : int}) # int i string oalrak da verebiliyorum fark etmiyor sonuç.
```

```python
df.dtypes
```

```python
df["Matches"] = df["Matches"].str.replace("*" ,"")

df['Matches'] = df['Matches'].astype(int)
```

Soru 2 : 10 yıldan uzun süre oynayn kriket oyuncularının Ortalama Batting_Strike_Rate değeri nedir ?

```python
df[df["Career_length"] > 10]["Batting_Strike_Rate"].mean() # çok güzel bir görüntü değil sanki..
```

```python
df.loc[df["Career_length"] > 10 , "Batting_Strike_Rate"].mean() # satırları filtreleyip ilgili sütuna ulaşıyoruz.sonrası kolay.
```

```python
# sorgu ile de yapılabilir.
df.query("Career_length > 10")["Batting_Strike_Rate"].mean()
```
